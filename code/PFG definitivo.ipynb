```python
# Defino una función para mostrar los valores en las barras de un gráfico de barras.
def show_values_on_bars(axs, h_v="v", space=0.4):
    """
    Esta función muestra los valores en las barras de un gráfico de barras.

    Parámetros:
    - axs: los ejes en los cuales se mostrarán los valores.
    - h_v: la orientación de las barras ('v' para vertical, 'h' para horizontal).
    - space: el espacio entre las barras y los valores mostrados.
    """
    # Defino una función interna para mostrar los valores en un solo gráfico
    def _show_on_single_plot(ax):
        """
        Esta función interna itera sobre las barras en los ejes dados y coloca
        el valor correspondiente en el centro o el extremo derecho de cada barra,
        dependiendo de la orientación especificada.
        """
        # Si la orientación es vertical
        if h_v == "v":
            # Itero sobre las barras en los ejes dados
            for p in ax.patches:
                # Calculo la posición x para centrar el valor
                _x = p.get_x() + p.get_width() / 2
                # Calculo la posición y para colocar el valor encima de la barra
                _y = p.get_y() + p.get_height()
                # Obtengo el valor de la altura de la barra
                value = int(p.get_height())
                # Coloco el valor centrado en la barra
                ax.text(_x, _y, value, ha="center")
        # Si la orientación es horizontal
        elif h_v == "h":
            # Itero sobre las barras en los ejes dados
            for p in ax.patches:
                # Calculo la posición x para colocar el valor a la derecha de la barra
                _x = p.get_x() + p.get_width() + float(space)
                # Calculo la posición y para colocar el valor alineado con la parte superior de la barra
                _y = p.get_y() + p.get_height()
                # Obtengo el valor del ancho de la barra
                value = int(p.get_width())
                # Coloco el valor a la derecha de la barra
                ax.text(_x, _y, value, ha="left")

    # Compruebo si se pasaron ejes únicos o un arreglo de ejes
    if isinstance(axs, np.ndarray):
        # Itero sobre los ejes en el arreglo
        for idx, ax in np.ndenumerate(axs):
            # Llamo a la función interna para mostrar los valores en el gráfico
            _show_on_single_plot(ax)
    else:
        # Llamo a la función interna para mostrar los valores en un solo gráfico
        _show_on_single_plot(axs)

# Defino una función similar a show_values_on_bars, pero con formateo adicional de los valores antes de mostrarlos.
def show_values_on_bars_formatted(axs, h_v="v", space=0.4):
    """
    Esta función es similar a show_values_on_bars, pero incluye un formateo adicional
    de los valores antes de mostrarlos.
    """
    # Defino una función interna para mostrar los valores en un solo gráfico
    def _show_on_single_plot(ax):
        """
        Esta función interna itera sobre las barras en los ejes dados y coloca
        el valor correspondiente en el centro o el extremo derecho de cada barra,
        después de aplicar el formateo.
        """
        # Si la orientación es vertical
        if h_v == "v":
            # Itero sobre las barras en los ejes dados
            for p in ax.patches:
                # Calculo la posición x para centrar el valor
                _x = p.get_x() + p.get_width() / 2
                # Calculo la posición y para colocar el valor encima de la barra
                _y = p.get_y() + p.get_height()
                # Obtengo el valor de la altura de la barra
                value = int(p.get_height())
                # Coloco el valor formateado centrado en la barra
                ax.text(_x, _y, formatter(value), ha="center")
        # Si la orientación es horizontal
        elif h_v == "h":
            # Itero sobre las barras en los ejes dados
            for p in ax.patches:
                # Calculo la posición x para colocar el valor a la derecha de la barra
                _x = p.get_x() + p.get_width() + float(space)
                # Calculo la posición y para colocar el valor alineado con la parte superior de la barra
                _y = p.get_y() + p.get_height()
                # Obtengo el valor del ancho de la barra
                value = int(p.get_width())
                # Coloco el valor formateado a la derecha de la barra
                ax.text(_x, _y, formatter(value), ha="left")

    # Compruebo si se pasaron ejes únicos o un arreglo de ejes
    if isinstance(axs, np.ndarray):
        # Itero sobre los ejes en el arreglo
        for idx, ax in np.ndenumerate(axs):
            # Llamo a la función interna para mostrar los valores en el gráfico
            _show_on_single_plot(ax)
    else:
        # Llamo a la función interna para mostrar los valores en un solo gráfico
        _show_on_single_plot(axs)

```

```python
# Defino una función llamada format_Attendance que toma un valor x como entrada.
def format_Attendance(x):
    # Convierto el valor a una cadena de caracteres para procesarlo.
    x = str(x)
    # Verifico si hay una coma en la cadena.
    if ',' in x:
        # Si hay una coma, elimino la coma y convierto la cadena a un número flotante.
        # Luego, redondeo el número y lo devuelvo.
        return round(float(x.replace(',', '')))
    else:
        # Si no hay una coma, simplemente convierto la cadena a un número flotante,
        # lo redondeo y lo devuelvo.
        return round(float(x))
```

```python
def correlated(df, threshold, target):
    '''Esta función devuelve las columnas de un dataframe cuya correlación con el objetivo está por encima del umbral'''
    # Filtrar solo columnas numéricas
    numeric_df = df.select_dtypes(include=['number'])
    # Calculo la matriz de correlación entre todas las columnas numéricas del dataframe y la columna del objetivo.
    corr = pd.Series(abs(numeric_df.corr()[target]) > threshold)
    # Extraigo las columnas cuya correlación absoluta con la variable objetivo es mayor que el umbral.
    corr_cols = corr.index[corr].tolist()
    # Remuevo la variable objetivo de la lista de columnas correlacionadas.
    corr_cols.remove(target)
    # Devuelvo la lista de columnas correlacionadas.
    return corr_cols

# Llamo a la función 'correlated' para encontrar las columnas del dataframe 'train_set' cuya correlación absoluta con la variable 'value' es mayor que 0.3.
num_cols_corr = correlated(train_set, 0.3, 'value')
```

```python
# Selecciono todas las columnas numéricas en el DataFrame train_set, excluyendo las columnas de tipo "object"
num_cols = train_set.select_dtypes(exclude="object").columns.tolist()

# Remuevo las columnas "value" y "Attendance" de la lista de columnas numéricas
num_cols.remove("value")
num_cols.remove("Attendance")
```

```python
# Elimino las columnas especificadas en la lista drop del DataFrame df_train
df_train = df_train.drop(drop, axis=1)

# Creo un DataFrame X que contiene las columnas numéricas y categóricas seleccionadas de df_train
X = df_train[num_cols + cat_cols]

# Creo un objeto y que contiene la columna "value" de df_train
y = df_train.value
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class Prepare_Categorical_Data(BaseEstimator, TransformerMixin):
    def __init__(self, cat_cols):
        self.cat_cols = cat_cols  # Inicializo la clase con una lista de columnas categóricas

    def fit(self, X, y=None):
        return self  # El método fit no hace nada en este caso, simplemente devuelve el objeto actual

    def transform(self, X):
        # Aplico las transformaciones necesarias a las columnas categóricas del DataFrame
        X['nationality'] = X["nationality"].map(format_nationality)  # Mapeo las nacionalidades
        X["nationality"] = X.nationality.map(lambda x: x if x in important_nations else "Other")  # Selecciono las nacionalidades importantes
        X.position.replace('0', None, inplace=True)  # Reemplazo los valores '0' en la columna 'position' por None
        X['position_1'] = X.position.map(format_position_1)  # Mapeo las posiciones a su primera letra
        return X[self.cat_cols]  # Devuelvo solo las columnas categóricas seleccionadas
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

# Defino un pipeline para el preprocesamiento de datos categóricos
cat_pipe = Pipeline([
    ('prepare_cat', Prepare_Categorical_Data(cat_cols)),  # Transformaciones personalizadas para datos categóricos
    ('imputer', SimpleImputer(strategy="most_frequent")),  # Imputación de valores faltantes
    ('onehot', OneHotEncoder(handle_unknown="ignore"))  # Codificación One-Hot de variables categóricas
])
```

```python
# Defino un pipeline para el preprocesamiento de datos numéricos
num_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy="mean")),  # Imputación de valores faltantes utilizando la media
    ('scaler', StandardScaler())  # Escalado de características numéricas para tener media cero y desviación estándar uno
])
```

```python
from sklearn.compose import ColumnTransformer

# Defino un ColumnTransformer para aplicar diferentes transformaciones a diferentes columnas del conjunto de datos
data_prep_pipe = ColumnTransformer([
    ("num", num_pipe, num_cols),  # Transformación para columnas numéricas
    ("cat", cat_pipe, cat_cols)   # Transformación para columnas categóricas
])
```

```python
def format_position_1(x):
    # Verifica si x no es None y si es una cadena (string)
    if x is not None and isinstance(x, str):
        # Comprueba si la longitud de x es mayor que 2
        if len(x) > 2:
            return x[:2]  # Devuelve los primeros dos caracteres de x
        else:
            return x  # Devuelve x si su longitud es menor o igual a 2
    else:
        return None  # Devuelve None si x es None o no es una cadena (string)
```

```python
# Aplico la transformación definida en el ColumnTransformer a las características de entrada X
X_prep = data_prep_pipe.fit_transform(X)
```

# MODELADO

```python
# Importo los módulos y clases necesarios
from sklearn.ensemble import RandomForestRegressor  # Importo el regresor de bosque aleatorio
from sklearn.model_selection import RandomizedSearchCV  # Importo la búsqueda aleatoria de hiperparámetros
from sklearn.svm import SVR  # Importo el regresor de vectores de soporte
import xgboost as xgb  # Importo la biblioteca XGBoost
import lightgbm as lgb  # Importo la biblioteca LightGBM
from scipy.stats import reciprocal, expon  # Importo distribuciones estadísticas para definir rangos de hiperparámetros
import joblib  # Importo joblib para guardar y cargar modelos y objetos de Python
```

```python
# Defino una función llamada get_model que devuelve un modelo de regresión basado en el nombre del modelo y sus parámetros
def get_model(model_name, params):
    if model_name == "xgb":  # Si el nombre del modelo es "xgb", creo un modelo XGBoostRegressor con los parámetros dados
        model = xgb.XGBRegressor(**params)
    elif model_name == "lgb":  # Si el nombre del modelo es "lgb", creo un modelo LGBMRegressor con los parámetros dados
        model = lgb.LGBMRegressor(**params)
    elif model_name == "svr":  # Si el nombre del modelo es "svr", creo un modelo SVR con los parámetros dados
        model = SVR(**params)
    else:  # De lo contrario, creo un modelo RandomForestRegressor con los parámetros dados
        model = RandomForestRegressor(**params)
    return model  # Devuelvo el modelo creado

# Defino una función llamada get_params que devuelve los parámetros adecuados según el nombre del modelo
def get_params(model_name):
    if model_name == "xgb":  # Si el nombre del modelo es "xgb", defino los parámetros para XGBoostRegressor
        params = {
            "n_estimators": [100, 200, 300, 400, 500],
            "max_depth": [3, 5, 7],
            "learning_rate": [1e-3, 1e-2, 1e-1, 0.5, 1.0],
        }
    elif model_name == "lgb":  # Si el nombre del modelo es "lgb", defino los parámetros para LGBMRegressor
        params = {
            "n_estimators": [100, 200, 300, 400, 500],
            "max_depth": [3, 5, 7],
            "learning_rate": [1e-3, 1e-2, 1e-1, 0.5, 1.0],
        }
    elif model_name == "svr":  # Si el nombre del modelo es "svr", defino los parámetros para SVR
        params = {
            'kernel': ['linear', 'rbf'],
            'C': reciprocal(20, 200000),
            'gamma': expon(scale=1.0),
        }
    else:  # De lo contrario, defino los parámetros para RandomForestRegressor
        params = {
            "n_estimators": [100, 200, 300, 400, 500],
            "max_depth": [3, 5, 7],
        }
    return params  # Devuelvo los parámetros definidos
```
